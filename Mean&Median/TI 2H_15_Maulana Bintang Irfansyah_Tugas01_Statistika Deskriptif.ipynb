```python
# Langkah pertama, meng-import 3 libraries python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
# Selanjutnya, mendefinisikan file yaitu dataset yang berekstensi "csv"
path="gdrive/MyDrive/Colab Notebooks/Statistik Komputasi/Data/data-pju-2017.csv"
df = pd.read_csv(path)
df
```

```python
# Kemudian, menentukan dataframe pada dataset, yaitu "Jumlah Titik PJ"
jumlah_pj_bandung = df['Jumlah Titik PJ']
```

```python
# Lalu, mencari mean dan median menggunakan fungsi mean dan median bawaan dari pyhton 
mean = jumlah_pj_bandung.mean()
median = jumlah_pj_bandung.median()
```

```python
# Berikutnya, menampilkan ukuran nilai tengah 
print('Mean: ', mean, 'nMedian: ', median)
```

```python
#@title
# Sehingga, pada langkah terakhir membuat grafik dengan library matplotlib untuk diketahuinya mean dan median dari dataset PJU Kota Bandung
plt.figure(figsize=(15, 5))
plt.hist(jumlah_pj_bandung, bins=35, color='grey')
plt.axvline(mean, color='red', label='Mean')
plt.axvline(median, color='yellow', label='Median')
plt.xlabel('Jumlah Titik PJ')
plt.ylabel('frekuensi')
plt.legend()
plt.show()
```
